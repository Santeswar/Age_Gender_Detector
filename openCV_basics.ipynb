```python
from cv2 import cv2
from matplotlib import pyplot as plt
```

# plotting

```python
img1 = cv2.imread("Resources/data/img1.jpg")
```

```python
plt.imshow(img1)
```

# arithmetic operations

```python
img2 = cv2.imread("Resources/data/img2.jpg")
plt.imshow(img2)
```

```python
result = img1+img2
plt.imshow(result)
# here both images should be of same size.
# similarly "-", "*", "/"
```

```python
img = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
figure_size = 7
result = cv2.blur(img, (figure_size, figure_size))
plt.figure(figsize=(11, 6))
plt.subplot(121), plt.imshow(cv2.cvtColor(img, cv2.COLOR_HSV2RGB)), plt.title("Original")
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(cv2.cvtColor(result, cv2.COLOR_HSV2RGB)), plt.title("Blurred")
plt.xticks([]), plt.yticks([])
plt.show()
```
